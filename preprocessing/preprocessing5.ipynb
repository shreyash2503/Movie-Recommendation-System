```python
import pandas as pd
import numpy as np
from tmdbv3api import TMDb
from tmdbv3api import Movie
from dotenv import load_dotenv
import os
import requests
from datetime import datetime
```

```python
load_dotenv()
```

```python
tmdb = TMDb()
tmdb.api_key = os.getenv('API_KEY')
tmdb.language = 'en'
tmdb.debug = True
```

```python
movie = Movie()
```

```python
data = pd.read_html("https://en.wikipedia.org/wiki/List_of_American_films_of_2022")
```

```python
names = []
```

```python
for i in range(2, 6):
    for i in data[i]['Title']:
        names.append(i)
```

```python
names = names[ : len(names) - 4]
```

```python
names
```

```python
movie_id = []
movie_not_found = []
```

```python
date_format = "%Y-%m-%d"
start_date = datetime.strptime("2022-01-01", date_format)
end_date = datetime.strptime("2022-12-31", date_format)

for name in names:
    print(name)
    try:
        search = movie.search(name)
    except Exception as e:
        print(e)
    
    if search:  # Check if the search result is not empty
        for res in search:
            if isinstance(res, str):
                print(res)
                break
            
            if res.get('release_date'):
                release_date = datetime.strptime(res['release_date'], date_format)
                if start_date <= release_date <= end_date:
                    movie_id.append((name, res['id']))
                    break
            else:
                print("Does not exist")
        else:
            print("Removed " + name + " from the names list")
            movie_not_found.append(name)
            # names.remove('name')
    else:
        print(f"No results found for {name}")
        movie_not_found.append(name)
        # names.remove('name')
```

```python
movie_id
```

```python
details = []
```

```python
def get_genres(obj):
    genres = []

    # print(type(obj))
    for i in obj:
        # print(i['name'])
        genres.append(i['name'])
    return genres
```

```python
for i in movie_id:

    url = f'https://api.themoviedb.org/3/movie/{i[1]}?api_key={os.getenv("API_KEY")}'
    response = requests.get(url)
    data_json = response.json()
    movie_details = {}
    movie_details['name'] = data_json['title']
    movie_details['id'] = i[1]
    movie_details['overview'] = data_json['overview'] 
    movie_details['genres'] = get_genres(data_json['genres'])
    details.append(movie_details)
```

```python
df = pd.DataFrame(details)
```

```python
df.head()
```

```python
def get_actors(id):
    url  =  f'https://api.themoviedb.org/3/movie/{id}/credits?api_key={os.getenv("API_KEY")}'
    resonse = requests.get(url)
    data_json = resonse.json()
    obj = data_json['cast']
    actors = []
    length = 5 if len(obj) > 5 else len(obj)
    for i in range(length):
        actors.append(obj[i]['name'])
    return actors
```

```python
def get_directors(id):
    url  =  f'https://api.themoviedb.org/3/movie/{id}/credits?api_key={os.getenv("API_KEY")}'
    resonse = requests.get(url)
    data_json = resonse.json()
    obj = data_json['crew']
    directors = []
    for i in obj:
        if(i['job'] == 'Director'):
            directors.append(i['name'])
            break
    return directors
```

```python
df['actors'] = df['id'].apply(get_actors)
```

```python
df.head()
```

```python
df['director'] = df['id'].apply(get_directors)
```

```python
df.head()
```

```python
def remove_space(obj):
    l = []
    for i in obj:
        l.append(i.replace(" ", ""))
    return l
```

```python
df['actors'] = df['actors'].apply(remove_space)
df['overview'] = df['overview'].apply(lambda x : x.split())
df['director'] = df['director'].apply(remove_space)
df['tags'] = df['overview'] + df['genres'] + df['actors'] + df['director']
df['tags'] = df['tags'].apply(lambda x : " ".join(x))
```

```python
df.head()
```

```python
new_df = df[['id', 'name', 'tags']]
```

```python
new_df
```

```python
df = new_df.rename(columns = {'id' : 'movie_id', 'name' : 'title'})
```
